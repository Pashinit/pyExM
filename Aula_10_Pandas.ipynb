Biblioteca Pandas

A utilização da biblioteca Pandas, uma biblioteca Python, para manipulação de dados. Oferece estruturas de dados e operações para manipular tabelas numéricas.
Esta biblioteca fornece duas estruturas de dados:

    -Series, um vetor de dados do mesmo tipo (um dimensão).
    -DataFrame, que contem um conjunto de Series, que podem ser de diferentes tipos (possui duas dimensões). Uma DataFrame pode ser vista como uma tabela de dados de uma base de dados ou uma folha de Excel.       
A estrutura mais utilizada na biblioteca Pandas é a aestrutura DataFrame.

O primeiro passo para trabaalhar com uma biblioteca no Python é importa-la com a utilização do comando import.

```python
import pandas as pd
```

Estrutura de dados

Série

Alguns exemplo de criação e manipulação de Series

```python
#Criacao de Series
#Criacao de uma serie a partir de um array numpy
import numpy as np
import pandas as pd
data = np.array(['a','b','c','d'])
s = pd.Series(data)
print('Serie 1: ')
print(s)
print('\n')
#Criacao de uma serie a partir de um array numpy, passandolhe os indices
data = np.array(['a','b','c','d'])
s = pd.Series(data,index=[100,101,102,103])
print('Serie 2: ')
print(s)
print('\n')
#Criacao de uma serie a partir de um dicionario
data = {'a':0.,'b':1.,'x':2.}
s = pd.Series(data)
print('Serie 3: ')
print(s)
print('\n')
print(s.ndim)
```

DataFrame

Alguns exemplo de criação de manipulação de DataFrames

```python
#Criacao de DataFrames
import pandas as pd
#Exemplo 1 - a partir de uma lista de valores
data = [1,2,3,4,5]
df = pd.DataFrame(data)
print('DataFrame 1: ')
print(df)
print('\n')

#Exemplo 2
data = [['Andre',9],['Matilde',11],['Goncalo',12]]
df = pd.DataFrame(data,columns=['Nomes','Idade'])
print('DataFrame 2: ')
print(df)
print('\n')

#Exemplo 3 - a partir de um dicionario
data = {'Nome':['Andre','Matidle'],'Idade':[9,10]}
df = pd.DataFrame(data)
print('DataFrame 3: ')
print(df)
print('\n')  

#Exemplo 4 - a partir de umalista de dicionarios - contem um missing value marcado como NaN
data = [{'a':1, 'b':2},{'a':5, 'b':10, 'c':20}]
df = pd.DataFrame(data)
print('DataFrame 4: ')
print(df)
print('\n')  

#Exemplo 5 - o exemlo de cima, mas dando titulos às linhas
data = [{'a':1, 'b':2},{'a':5, 'b':10, 'c':20}]
df = pd.DataFrame(data,index=['linha1','linha2'])
print('DataFrame 5: ')
print(df)
print('\n') 
```

Operações sobre DataFrames:

    1.Manipular colunas (mostrar, adicionar, apagar)
    2.Manipular linhas (mostrar, adicionar, apagar)

```python
#Mostrar so uma coluna
import pandas as pd
print('Manipular colunas: ')
data = {'col1':pd.Series([1,2,3], index=['a','b','c']),'col2':pd.Series([1,2,3,4], index=['a','b','c','d'])}
df = pd.DataFrame(data)
print(df['col1'])
print('\n') 
#Adicionar uma coluna a um DataFrame existente
data = {'col1':pd.Series([1,2,3], index=['a','b','c']),'col2':pd.Series([1,2,3,4], index=['a','b','c','d'])}
df = (pd.DataFrame(data))

print("Adicionar uma nova coluna do tipo Serie: ")
df['col3']=pd.Series([10,20,30],index=['a','b','c'])
print(df)
print('\n') 

print("Adicionar uma nova coluna do tipo Serie que soma as dias colunas existentes: ")
df['col4']=df['col1']+df['col3']
print(df)
print('\n') 

#Apagar um coluna
d = {'col1':pd.Series([1,2,3], index=['a','b','c']),
     'col2':pd.Series([1,2,3,4], index=['a','b','c','d']),
     'col3':pd.Series([10,20,30], index=['a','b','c'])}

df = pd.DataFrame(d)
print('DataFrame Original: ')
print(df)
print('\n') 

#utilizando a funcao del
print("Apagar a primeira coluna: ")
del df['col1']
print(df)
print('\n') 

#utilizando a fincao pop
print("Apagar a segunda coluna com pop: ")
df.pop('col2')
print(df)
```

```python
import pandas as pd
#Aceder a um alinha
data = {'linha1':pd.Series([1,2,3], index=['a','b','c']),
        'linha2':pd.Series([1,2,3,4], index=['a','b','c','d'])}
print('DataFrame data: ')
print(data)
print('\n') 

df = pd.DataFrame(data)
print("Retirar os valores na linha com indice 2 (terceira linha): ")
print(df.iloc[2])
print('\n') 

#Mostrar varias linhas
print('Mostrar as linhas com os indices 2 e 3: ')
print(df[2:4])
print('\n') 

#Adicionar linhas
df1 = pd.DataFrame([[1,2],[3,4]], columns = ['a','b'])
df2 = pd.DataFrame([[5,6],[7,8]], columns = ['a','b'])
print("Adicionar as linhas de df2 a df1: ")
df1 = pd.concat([df1,df2],ignore_index=True)
print(df1)
print('\n') 
#Apagar linhas
print("Apagar linha com indice 0")
df1 = df1.drop(0) #Apaga as linha scom indice 0
print(df1)
```

1.Crie um dicionario de 4 perguntas e respostas, diga a media e o numero de perguntas certas

```python
import pandas as pd
perguntas = {
    'Pergunta 1':{
        'perguntas': 'Quanto é 2+2',
        'respostas': {'a':'1', 'b':'4', 'c':'11'},
        'resposta_certa':'b',
    },
    'Pergunta 2':{
        'perguntas': 'Quanto é 3*2',
        'respostas': {'a':'6', 'b':'4', 'c':'10'},
        'resposta_certa':'a',
    },
    'Pergunta 3':{
        'perguntas': 'Quanto é 3-2',
        'respostas': {'a':'6', 'b':'11', 'c':'1'},
        'resposta_certa':'c',
    },
    'Pergunta 4':{
        'perguntas': 'Quanto é 3%2',
        'respostas': {'a':'2', 'b':'1', 'c':'0'},
        'resposta_certa':'b',
    }
}
```

```python
df = pd.DataFrame(perguntas)
ac = 0

for c in range (len(df)+1):
    for v in range (len(df)-1):
        print(df.iloc[v,c])
    x = str(input("Resposta: "))
    if (x == df.iloc[v+1,c]):
        print("Resposta certa!")
        ac += 1
    else:
        print("Resposta errada!")
    print("")
print("")
print("Respostas certas: ",ac)
print("Média de respostas certas: ",ac/(len(df)+1))
```
